```python
import os
import shutil

def delete_folder(path):
    if os.path.exists(path) and os.path.isdir(path):
        print(f"Deleting: {path}")
        shutil.rmtree(path)
        print("✔ Deleted")
    else:
        print(f"✖ Not found or not a directory: {path}")

# === Use current working directory for Jupyter ===
base_dir = os.getcwd()  # Automatically handles notebooks
print("base_dir: ", base_dir)

# === List of folder names to delete (relative to base_dir) ===
folders_to_delete = [
    "checkpoints_v72_baseline_W336_D12",
    "checkpoints_hierarchical_20251030_231912",
    "checkpoints_hierarchical_stable_20251101_122609",
    "checkpoints_hierarchical_stable_20251101_122659",
    "checkpoints_ssm_d384_l8_bs32_e30_20251031_120854",
    "documentation v1_0 hierarchical",
    "placeholder",
    "placeholder"
]

# === Delete loop ===
for folder in folders_to_delete:
    full_path = os.path.join(base_dir, folder)
    delete_folder(full_path)


print("Complete.")
```
